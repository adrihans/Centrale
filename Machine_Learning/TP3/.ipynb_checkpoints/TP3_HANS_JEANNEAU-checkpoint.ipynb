## 1. Logistic regression, IRLS algorithm.

### Preliminary question: the algorithm

#### Have a look at the function `regression_logistique.m` and locate the main steps of the algorithm you have been taught (see course).
You can comment the code in the cell below.

```python
def regression_logistique(X,t,Nitermax=20,eps_conv=1e-3):
    '''Entrees :
    X = [ones(N_train,1) x_train];
    t = class_train 
    Nitermax = nombre maximale d'itérations (20 par défaut)
    eps_conv = critère de convergence sur norm(w-w_old)/norm(w) ; 
    eps_conv=1e-3 par défaut
    
    Sorties : 
    w : vecteur des coefficients de régression logistique
   Niter : nombre d'itérations utilisées effectivement
   
   Fonction de régression logistique pour la classification binaire.
   
   Utilisation :
       Nitermax = 50
       eps_conv = 1e-4
       [w,Niter] = regression_logistique(X,t,Nitermax,eps_conv)
    '''
    N_train = X.shape[0]

    #initialisation : 1 pas de l'algorithme IRLS
    w = np.zeros((X.shape[1],))
    w_old = w 
    y = 1/2*np.ones((N_train,))
    R = np.diag(y*(1-y))   # diag(y_n(1-y_n))
    z = X.dot(w_old)-la.inv(R).dot(y-t)
    w = la.inv(X.T.dot(R).dot(X)).dot(X.T).dot(R).dot(z)

    # boucle appliquant l'algortihme de Newton-Raphson
    Niter = 1
    while ( (la.norm(w-w_old)/la.norm(w)>eps_conv) & (Niter<Nitermax) ):
        Niter = Niter+1
        y = 1/(1+np.exp(-X.dot(w)))
        R = np.diag(y*(1-y))  
        w_old = w 
        z = X.dot(w_old)-la.inv(R).dot(y-t) 
        w = la.inv(X.T.dot(R).dot(X)).dot(X.T).dot(R).dot(z)
         
    return w, Niter
```

### Reading & preparing synthetic data

Load the training and test data sets `synth_train.txt`
and `synth_test.txt`. The targets t belong to {1,2} and the features  
x belong to R^2. 

We have 100 training samples and 200 test samples

* the 1st column contains the label of each sample, 
* columns 2 and 3 contain the coordinate of each point.

```python
# Training set
synth_train = np.loadtxt('synth_train.txt') 
class_train = synth_train[:,0]
class_train_1 = np.where(synth_train[:,0]==1)[0]
class_train_2 = np.where(synth_train[:,0]==2)[0]
x_train = synth_train[:,1:]
N_train = np.size(x_train,axis=0)

# Test set
synth_test = np.loadtxt('synth_test.txt')
class_test = synth_test[:,0]
class_test_1 = np.where(synth_test[:,0]==1)[0]
class_test_2 = np.where(synth_test[:,0]==2)[0]
x_test = synth_test[:,1:]
N_test = np.size(x_test,axis=0)
```

### Preparing features for logistic regression (binary classification)
First, we prepare the feature matrix and the target vector associated to 
the training and test sets:


```python
X = np.hstack((np.ones((N_train,1)),x_train))
t = 2-class_train   # 0 if class=2, 1 if class=1

X_test = np.hstack((np.ones((N_test,1)),x_test))
t_test = 2-class_test   # 0 if class=2, 1 if class=1
```

1. $ $

```python
w, Niter = regression_logistique(X, t, Nitermax=50, eps_conv=1e-3)
```

2.  $y_n=\sigma(w^T x)$, with $\sigma(x) = \frac{1}{1+e^{-x}}$

```python
def sigmoid(x):
    return (1 + np.exp(-np.array(x)))**-1

y = sigmoid(w.T.dot(X_test.T))
```

3. We display the test set points colored according to the ground truth as well as the decision boundary  $w^Tx = 0$, to demonstrate its accuracy: 

```python
# First compute w... then: 
colors = ['#28E0A8', '#0B70E8']

fig = plt.figure(figsize=(8, 8))

x1 = np.linspace(-2.5,1.5,10) 
x2 = (-w[0]-w[1]*x1)/w[2]
plt.plot(x1,x2,'k--')
plt.scatter(X_test[:, 1], X_test[:, 2], c=[colors[int(t)] for t in t_test], label='0')
plt.scatter([], [], c=colors[1], label=1)
_ = plt.legend()
_ = plt.title('Logistic regression applied to our dataset')
plt.show()
```

We can see on this 2-dimensional example that the decision boundary seems to be a good choice

4. $y_n=\sigma(w^T x) > \frac{1}{2}$

5. We create here 5 additional points that does not follow the data distribution, being away from the clusters and sometimes on the wrong side of the boundary. We add them to the dataset and apply logistic regression to this new dataset. We can then compute the boundary of this model and compare it to the boundary found on the regular dataset.

```python
# Create additional outliers
additionalpoints = np.array([[1, -5, 0], [1, 3, 3], [1, 0, -2], [1, 0, 5], [1, 5, 5]])
additionallabels = np.array([1, 1, 1, 0, 0])

# Add them to the train set
X2 = np.hstack((np.ones((N_train,1)),x_train))
X2 = np.vstack((X2,  additionalpoints))
t2 = 2-class_train   # 0 if class=2, 1 if class=1
t2 = np.hstack((t, additionallabels))
```

```python
# Apply logistic regression to the new train set
w2, Niter = regression_logistique(X2, t2, Nitermax=50, eps_conv=1e-3)

fig=plt.figure(figsize=(8, 8), dpi= 80)

# Plot former boundary
x1 = np.linspace(-4, 3,10) 
x2 = (-w[0]-w[1]*x1)/w[2]
plt.plot(x1,x2,'g--')

# Plot new boudary
x1_prime = np.linspace(-3, 1.5, 10) 
x2 = (-w2[0]-w2[1]*x1_prime)/w2[2]
plt.plot(x1_prime,x2,'r--')

# Plot test points and outliers
colors = ['#28E0A8', '#0B70E8']

colors2 = ['#289058', '#0B0098']
plt.scatter(X_test[:, 1], X_test[:, 2], c=[colors[int(t)] for t in t_test])
plt.scatter(additionalpoints[:, 1], additionalpoints[:, 2], c=[colors2[int(t)] for t in additionallabels])
```

The green dashed line represent the decision boundary of the logistic regression applied to the regular dataset, and the red one represent the decision boundary of the logistic regression applied to the dataset with outliers. The darker dots are the outliers

We can notice that these 5 outliers (on a dataset of 100) significantly affect the boundary. The new boundary is shifted from its original position, sufficiently to become less efficient as a decision boundary but is still relevant for classification.

```python
train_5 = np.loadtxt('train_5.txt',delimiter=',')   # 556 samples
train_6 = np.loadtxt('train_6.txt',delimiter=',')   # 664 samples
```

### Separating the training and test sets

We keep in the training set the 145 first images of 5s and the 200 first
images of 6s:


```python
x_train_brut = np.vstack((train_5[:145,:], train_6[:200,:]))
N_train = np.size(x_train_brut,axis=0)
class_train = np.ones((345,))   # label 1 for digit 6
class_train[:145] = 0       # label 0 for digit 5

x_test_brut = np.vstack((train_5[145:,:], train_6[200:,:]))
N_test = np.size(train_5,axis=0)+np.size(train_6,axis=0)-N_train
```

2. We compute the standard deviation for each 256 pixel, across all the 345 train examples. There are 173 pixels having a standard deviation above 0.5, which will be extracted as our new data representation dimension 

```python
stds = np.array([x_train_brut[:, i].std() for i in range(256)])
lis_sig = np.argwhere(stds > 0.5)[:,0]
print('there are {} significant pixels'.format(lis_sig.shape[0]))
```

3. We can visualize the pixels with a standard deviation above 0.5. The ones below are located on the right and left edges, which makes sense since 5 and 6 tends to occupy a vertical rectangular area, leaving the outside pixels often black 

```python
print('3.\nHere are the significant pixels')
sig_mask = np.array([stds > 0.5]).reshape((16, 16))
plt.imshow(sig_mask, cmap='gray')
plt.show()
```

4. and 5. We extract the significant pixels as our new train and test sets. We also create our ground truth labes for the test set because it hasn't been created previously

```python
x_train = x_train_brut[:, lis_sig]
x_test = x_test_brut[:, lis_sig]

X_train = np.hstack((np.ones((345,1)),x_train))
X_test = np.hstack((np.ones((N_test,1)),x_test))
```

```python
# Create labels for test set
class_test = np.ones((N_test,))
class_test[:train_5.shape[0]-145] = 0
```

6. We can then apply logistic regression to our problem and use it to make predictions (7.) and get the error rate (8.)

```python
w, Niter = regression_logistique(X_train, class_train, Nitermax=13, eps_conv=1e-3)
```

7. 

```python
out = sigmoid(X_test.dot(w.T))
pred = out > 0.5
```

8. 

```python
erreur = sum(abs(pred-class_test))/N_test
print('Classification error rate: {0:.2%}'.format(erreur))
```

9. Let's visualize some misclassified examples, three of each class

```python
n = 0

def plot_nth_error(n):
    error_index = np.argwhere(pred != class_test)[n, 0]

    n5 = train_5.shape[0] - 145
    if error_index < n5:
        I = train_5[error_index]
    else:
        I = train_6[error_index-n5]

    plt.imshow(I.reshape(16, 16) ,cmap='gray')

plt.subplot(331); plot_nth_error(0)
plt.subplot(332); plot_nth_error(1)
plt.subplot(333); plot_nth_error(2)
plt.subplot(334); plot_nth_error(66)
plt.subplot(335); plot_nth_error(67)
plt.subplot(336); plot_nth_error(68)
```

Some of these misclassified examples have exotic shapes, such as the 2nd $6$. Some others have a more regular shape but are distorted, such as the 3rd $5$ which is flatten. Some also occupy pixels that have been dropped out of the train and test sets, such as the 3rd $6$.

It could be interesting to look at the *worst classified examples*, i.e examples that have been classified with the greates level of certainty as the opposite class.

```python
def plot_n_worst_errors(N):
    errors_index = np.argwhere(pred != class_test)[:, 0]
    distance_to_target = np.abs(out[errors_index] - class_test[errors_index])
    N_worst_predictions_index = distance_to_target.argsort()[-N:][::-1]
    N_worst_predictions = errors_index[N_worst_predictions_index]
    
    n5 = train_5.shape[0] - 145
    for n in N_worst_predictions:
        if n < n5:
            I = train_5[n]
        else:
            I = train_6[n-n5]

        plt.imshow(I.reshape(16, 16) ,cmap='gray')
        plt.show()
        
plot_n_worst_errors(3)
```

These examples have been classified as the opposite class with the greatest degree of certainty. For instance, the 2nd is a 5 but looks a lot like a 6, only 2 pixels are preventing the lower loop to close. The 3rd also has the general shape of a 6 while being a 5.

Even if logistic regression has a decent score in this situation, with an error rate below 8%, its use as a character recognition model is quite questionable. First of all, because it is not translation-invariant nor scale-invariant. shifting the whole image by 1 pixel vertically or horizontally would mean completely changing the input variables, which would have a great impact on the output. This model is also unable to grasp non-linear relationship between pixels, which is a con.

1. and 2. Let's use the logistic regression implementation of scikit-learn with default parameters, i.e a l2 regularization and a lambda of 1. The default score implemented for logistic regressions is the accuracy, so in order to compare it with the error rate studied previously we will compute $1-Accuracy = ErrorRate$

```python
from sklearn.linear_model import LogisticRegression

logistic_regression = LogisticRegression(random_state=42, solver='lbfgs') 

logistic_regression.fit(x_train, class_train)
error_rate = 1 - logistic_regression.score(x_test, class_test)
print('Error rate of sklearn logistic regression with default parameters: {:.2%}'.format(error_rate))
```

The results are quite superior to those obtained with our method.

We can now try different parameters and see if we will get an improvement of the score. To do so, we will use a gridsearch that will logarithmically explore values of $C=\lambda^{-1}$ between $0.1$ and $10$, for both L1 and L2 regularizations

```python
from sklearn.model_selection import GridSearchCV

logistic_regression = LogisticRegression(random_state=42, solver='liblinear') 
parameters = {'penalty': ['l2', 'l1'],
              'C': [0.01, 0.02, 0.05, 0.1, 0.2, 0.5, 1, 2, 5]}

clf = GridSearchCV(logistic_regression, parameters, cv=3)
clf.fit(x_train, class_train)
```

```python
error_rate = 1 - clf.score(x_test, class_test)
C = clf.best_estimator_.get_params()['C']
penalty = clf.best_estimator_.get_params()['penalty']
print('Error rate of sklearn logistic regression with gridsearch on C and : {:.2%}, for C = {} and a {} penalty'.format(error_rate, C, penalty))
```

We obtain a slightly better score, which means that the default parameters of the logistic regression were not the best but close to optimal. This is often not the case, so it is a good practice to explore the hyperparameters space to find the optimal hyperparameters for the given situation. The best hyperparameters obtained by the gridsearch are C=0.05 for the L2 penalty.

To conclude, Logistic Regression is a classical model, simple but robust and can be sufficient depending on the situation. It has however limitations, as it cannot grasp non-linear patterns and is not shift-invariant. For image analysis, Convolutional Neural Networks would be more appropriate.
