### 4)

```python
def split_train_test(Y, c,k=5):
    n = Y.shape[0]
    Y_split = np.array([Y[int(i*n/k):int((i+1)*n/k)] for i in range(0,k)]) 
    c_split = np.array([c[int(i*n/k):int((i+1)*n/k)] for i in range(0,k)])
    return Y_split, c_split

Y_split, c_split = split_train_test(Y, c)
```

```python
np.array(Y_split).shape
```

Nous retrouvons bien les $5*60=300$ elements de dimension 2 du dataset

### 5)

Pour augmenter la lisibilité du code, nous avons créé une fonction annexe regroupant les $K=5$ split en train et test set en mettant de côté un des split pour le test set et les autres pour le train set.

```python
def get_split_i(Y_split, c_split, i):
    nb_split, split_size, N = Y_split.shape

    Y_train = np.append(Y_split[0:i], Y_split[i:4], axis=0)
    Y_train = Y_train.reshape((nb_split-1)*split_size, N)
    Y_test = Y_split[i]
    
    c_train = np.append(c_split[0:i], c_split[i:4], axis=0)
    c_train = c_train.reshape((nb_split-1)*split_size)
    c_test = c_split[i]
    return Y_train, Y_test, c_train, c_test
```

Ensuite, nous itérons sur les K manières d'arranger les train et test set et entrainons les différent modèles que nous stockons, avec les scodes associés, dans le dictionnaire `results`

Nous produisons également le classifieur de fusion par vote majoritaire demandé à la question $7)$

```python
Y_split, c_split = split_train_test(Y, c)
results = {'DTC': {'models': [], 'scores': []},
           'GNB': {'models': [], 'scores': []},
           'KNC': {'models': [], 'scores': []},
           'Fusion': {'models': [], 'scores' : []}}

for i in range(0,5):
    Y_train, Y_test, c_train, c_test = get_split_i(Y_split, c_split, i)
    
    #DecisionTreeClassifier
    DTC=DecisionTreeClassifier(max_depth=5)
    DTC.fit(Y_train, c_train)

    #GaussianNB
    GNB = GaussianNB()
    GNB.fit(Y_train, c_train)

    #KNeighborsClassifier
    KNC=KNeighborsClassifier(n_neighbors=1)
    KNC.fit(Y_train, c_train)

    #Fusion 
    Fusion=VC([('DTC', DTC),('GNB', GNB),('KNC',KNC)])
    Fusion.fit(Y_train, c_train)

    for name, model in [('DTC', DTC),('GNB', GNB),('KNC',KNC), ('Fusion', Fusion)]:
        results[name]['models'].append(model)
        results[name]['scores'].append(model.score(Y_test, c_test))
```

### 6)

Nous implementons maintenant les fonctions `frontieres` et `plot_results` permettant d'afficher les resultats, que nous utilisons ensuite.

```python
def frontieres(clf, X=Y):
    h = .02
    x1_min, x1_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    x2_min, x2_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    x11, x22 = np.meshgrid(np.arange(x1_min, x1_max, h),np.arange(x2_min, x2_max, h))
    if hasattr(clf, "decision_function"):
        Z = clf.decision_function(np.c_[x11.ravel(), x22.ravel()])
    else:
        Z = clf.predict(np.c_[x11.ravel(), x22.ravel()])
    Z=Z.reshape(x11.shape)
    plt.contourf(x11, x22, Z, cmap=plt.cm.bwr, alpha=.8)
```

### 7) 

```python
def plot_results(results, Y_split, c_split):
    fig = plt.figure(figsize=(15, 15))
    models = list(results.keys())
    N = len(models)
    M = len(results[models[0]]['scores'])
    
    for i in range(M):
        plt.subplot(M, N+1, (N+1)*i+1)
        plt.scatter(Y_split[i,:,0], Y_split[i,:,1], c=c_split[i])
        for j in range(N):
            plt.subplot(M, N+1, (N+1)*i+j+2)
            frontieres(results[models[j]]['models'][i])

    plt.show()
```

```python
plot_results(results, Y_split, c_split)
```

Nous pouvons observer les différences de comportement des modèles choisis.
* La `Random Forest` de profondeur maximale égale à 5 a pour frontières de décision des droites colinéaires aux deux dimensions, ce qui est performant mais non suffisant dans le cas de nos données a la frontière idéale non-linéaire.
* Le `Naive Bayes Classifier` a une frontière de décision quasi-linéaire
* Le `K  Nearest Neighbors` à 1 seul voisin a naturellement tendance a overfitter fortement

Le modèle de Fusion a Vote Majoritaire a, quant à lui, une frontière de déciion qui correspond a ce qu'un humain aurait pu vouloir dessiner. La frontière combine la précision du KNN et la régularité des autres modèles.

### 8)

```python
def plot_scores(results):
    scores_matrix = []
    models=list((results.keys()))
    for model in models:
        scores_matrix.append(results[model]['scores'])
    scores_matrix = np.array(scores_matrix).T
    sns.heatmap(scores_matrix, annot=True, xticklabels=models)
    plt.show()
    
plot_scores(results)
```

En affichant les scores des différents modèles, on se rend compte d'un fait surprenant: le KNN a des performances égales à 1 sur la majorité des plis ! Les train et test sets ne sont pas mélangés (cela affecterai un KNN plus que tout autre modèle), cela doit donc provenir du nombre de voisins k=1. Au vu des données, il est possible que le plus proche voisin soit un très bon indicateur de la classe.

Dans un cas où il n'y aurait pas de modèle presque parfait parmi les modèlles a fusionner, fusionner les modèles devrait améliorer les performances globales en donnant plus de robustesse et en combinant les avantages des différents modèles.

### Exercice II
### 1) 2) et 3)

Pour chaque pli train - test, nous voulons obtenir les performances de chaque modèle afin de les utiliser comme poids pour la création d'une Fusion a Vote Majoritaire Pondéré. Afin d'augmenter la robustesse des modèles, nous voulons étendre la cross-validation a un second niveau et entrainer K fois les modèles sur des splits du train set (i.e un train_small et un validation set)

Nous avons implémenté cette double cross-validation comme fonction pouvant utiliser une fonction softmax pour mieux différentier les modèles, afin de préparer a la question 4) et eviter les redondances

```python
def softmax(x, nu):
    x*=nu
    return np.exp(x - max(x)) / np.sum(np.exp(x - max(x)))
```

```python
def double_cv(Y_split, c_split, K, nu=None):
    K = 5
    Y_split, c_split = split_train_test(Y, c, K)
    results = {'DTC': {'models': [], 'scores': []},
               'GNB': {'models': [], 'scores': []},
               'KNC': {'models': [], 'scores': []},
               'DTC2': {"models": [], "scores": []},
               'FusionM': {'models': [], 'scores' : []},
               'FusionMP': {'models': [], 'scores' : []}}

    for i in range(K):
        Y_train, Y_test, c_train, c_test = get_split_i(Y_split, c_split, i)

        W = np.zeros(4)

        Y_train_split, c_train_split = split_train_test(Y_train, c_train, K)
        for j in range(K):
            Y_train_small, Y_val, c_train_small, c_val = get_split_i(Y_train_split, c_train_split, j)
            #DecisionTreeClassifier
            DTC = DecisionTreeClassifier(max_depth=5)
            DTC.fit(Y_train, c_train)
            #GaussianNB
            GNB = GaussianNB()
            GNB.fit(Y_train, c_train)
            #KNeighborsClassifier
            KNC = KNeighborsClassifier(n_neighbors=1)
            KNC.fit(Y_train, c_train)
            #DecisionTreeClassifier2
            DTC2 = DecisionTreeClassifier(max_depth=1)
            DTC2.fit(Y_train, c_train)
            #Add model scores to the weights
            clf = [DTC, GNB, KNC, DTC2]
            for p in range(4):
                W[p] += 1/K * clf[p].score(Y_val, c_val)
        if nu is not None:
            W = softmax(W, nu)

        #DecisionTreeClassifier
        DTC=DecisionTreeClassifier(max_depth=5)
        DTC.fit(Y_train, c_train)

        #GaussianNB
        GNB = GaussianNB()
        GNB.fit(Y_train, c_train)

        #KNeighborsClassifier
        KNC=KNeighborsClassifier(n_neighbors=1)
        KNC.fit(Y_train, c_train)

        #DecisionTreeClassifier2
        DTC2 = DecisionTreeClassifier(max_depth=1)
        DTC2.fit(Y_train, c_train)

        #Fusion Majoritaire 
        FusionM=VC([('DTC', DTC), ('GNB', GNB), ('KNC',KNC), ('DTC2', DTC2)])
        FusionM.fit(Y_train, c_train)

        #Fusion Majoriataire Pondérée
        FusionMP=VC([('DTC', DTC), ('GNB', GNB), ('KNC',KNC), ('DTC2', DTC2)], weights=W)
        FusionMP.fit(Y_train, c_train)

        for name, model in [('DTC', DTC),('GNB', GNB),('KNC',KNC), ('DTC2', DTC2),
                            ('FusionM', FusionM), ('FusionMP', FusionMP)]:
            results[name]['models'].append(model)
            results[name]['scores'].append(model.score(Y_test, c_test))
    return results
```

```python
results = double_cv(Y_split, c_split, 5)
```

```python
plot_results(results, Y_split, c_split)
```

```python
plot_scores(results)
```

Nous pouvons maintenant comparer les fusions a vote majoritaire pondérées et non-pondérées. Nous remarquons que la seconde a une frontière de décision plus proche de ce que nous recherchons et a de meilleures performances que la première sur chaque pli.

Nous utilisons maintenant une fonction softmax afin de mieux differencier les modèles

```python
results = double_cv(Y_split, c_split, 5, nu=15)
```

```python
plot_results(results, Y_split, c_split)
```

```python
plot_scores(results)
```

Avec une valeur de $\eta$ élevée, la méthode de vote tends vers une séléction du modèle au meilleure score. En effet, dans la fonction softmax, le meilleure score a tendance a "écraser" les autres de par l'exponentielle et reste la seule choisie.

```python
results = double_cv(Y_split, c_split, 5, nu=2)
```

```python
plot_scores(results)
```

En essayant une valeur de $\eta$ plus basse, nous trouvons un compromis raisonnable permetatnt de mieux mettre en avant les modèles performants sans pour autant réduire l'influence des autres modèles à zero.
