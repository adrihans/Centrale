## Question 2 : 

*Transformez le classifieur faible du TP n°1 ("Bad Tree") en classifieur fort grâce à l’algorithme AdaBoost.
Chaque membre du comité est donc un arbre de décision de profondeur maximale 1 (tree stump) et est une
instance de DecisionTreeClassifier(max_depth=1). On prendra M = 20 classifieurs dans le comité.
Pour qu’un arbre minimise la perte pondérée par les poids $w_m^{(i)}$ d’AdaBoost, vous devrez obtenir un échantillon bootstrap à partir de vos données généré en 1). Cet échantillon s’obtient en tirant au hasard une donnée avec la probabilité: $$p_i=\frac{w_m^{(i)}}{\sum_{i'=1}^nw_m^{(i')}}$$*


*On pourra utiliser `np.random.multinomial` à cette fin.*

*Comment évolue le risque empirique sur les données d’entraînement ?*


```python
#Classifieur faible : 
#Arbre de décision de profondeur maximale 1
maxdepth=1
#number of classifiers :
M=20 

def Adaboost(Y,c,maxdepth,M):
    n=len(c)
    #On se replace dans le cas binaire {-1,1} : 
    c=np.array(c*2-1)
    alphas=[]
    models=[]
    accuracy=[]  
    epsilons=[]
    c_pd = pd.DataFrame(c.copy())
    c_pd['weights'] = 1/n # On attribue les poids initiaux : w = 1/n
        
    for k in range(1, M+1):
        predictions=[]
        #Sampler un dataset :        
        #Il faut prendre la taille = 4*n
        Dboost=np.random.multinomial(n=n,pvals=(c_pd['weights']/(c_pd['weights'].sum())),size=4).reshape(4*n) 
        #On définit Yboost :
        Yboost=np.array([Y[i] for i in Dboost]).reshape(4*n,2)
        #On définit les classes associées cboost :
        cboost=np.array([c[i] for i in Dboost])
        cboost_pd = pd.DataFrame(cboost.copy())
        cboost_pd['weights'] = [c_pd['weights'][i] for i in Dboost]
        
        
        #On défnit l'arbre :
        DTC = DecisionTreeClassifier(max_depth=maxdepth)
        #On entraine l'arbre :
        model=DTC.fit(Y,c, sample_weight=np.array(c_pd['weights']))
        #On stock le modèle :
        models.append(model)
        #Prediction du model :
        c_pred=model.predict(Y)
        #On calcule epsilon:
        c_pd['misclassified'] = np.where(c_pred != c ,1,0)
        epsilon = np.sum(c_pd['weights']*c_pd['misclassified'])/np.sum(c_pd['weights'])
        #On calcule alpha:
        alpha=np.log((1-epsilon)/epsilon)
        alphas.append(alpha)
        #Mise à jour du poids :
        c_pd['weights'] = c_pd['weights']*np.exp(-0.5*alpha*c_pred*c)
        
    return alphas, models
 
def Adaboost_predict(alphas, models, Y):
    predictions=[]
    for alpha, model in zip(alphas, models):
        # We use the predict method for the single decisiontreeclassifier models in the list
        prediction = alpha*model.predict(Y) 
        predictions.append(prediction)

    #On fait la prediction finale :
    final_pred=np.sign(np.sum(np.array(predictions), axis=0))    
    return final_pred

def Adaboost_accuracy(alphas, models, Y, c):
    accuracy=[]  
    n=len(c)
    #On se replace dans le cas binaire {-1,1} : 
    c=np.array(c*2-1)
    M=len(models)
    for i in range(M):
        final_pred=Adaboost_predict(alphas[:i+1], models[:i+1], Y)
        accuracy.append(np.sum(np.array(final_pred)==c)/(n))
    return accuracy
```

```python
alphas, models=Adaboost(Y,c,maxdepth,M)
accuracy=Adaboost_accuracy(alphas, models, Y, c)
```

**On affiche l'évolution du risque empirique sur les données d'entrainement :**

```python
fig=plt.figure(figsize=(15,7))
plt.plot([k for k in range(1,M+1)], accuracy)
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.title('Accuracy with iteration :')
plt.show()
```

**Même si certaines oscillations peuvent être présentes, plus le nombre de classifieurs dans le comité est important meilleurs sont les résultats de classification.**

**Ce résultat était évidemment attendu et justifie l'utilisation d'une telle méthode.**

### question 4 :
*Affichez l’évolution du comité au fil des itérations d’AdaBoost ainsi que les arbres successivement appris.*

**Nous implementons maintenant les fonctions `frontieres` et `frontieres_Adaboost` permettant d'afficher les resultats, que nous utilisons ensuite.**

```python
def frontieres(clf, X=Y):
    h = .02
    x1_min, x1_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    x2_min, x2_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    x11, x22 = np.meshgrid(np.arange(x1_min, x1_max, h),np.arange(x2_min, x2_max, h))
    if hasattr(clf, "decision_function"):
        Z = clf.decision_function(np.c_[x11.ravel(), x22.ravel()])
    else:
        Z = clf.predict(np.c_[x11.ravel(), x22.ravel()])
    Z=Z.reshape(x11.shape)
    plt.contourf(x11, x22, Z, cmap=plt.cm.bwr, alpha=.8)
```

```python
def frontieres_Adaboost(alphas, models, X=Y):
    h = .02
    x1_min, x1_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    x2_min, x2_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    x11, x22 = np.meshgrid(np.arange(x1_min, x1_max, h),np.arange(x2_min, x2_max, h))
    Z = Adaboost_predict(alphas, models, np.c_[x11.ravel(), x22.ravel()])
    Z=Z.reshape(x11.shape)
    plt.contourf(x11, x22, Z, cmap=plt.cm.bwr, alpha=.8)
```

```python
#Génération du dataset :
(Y,c) = make_moons(n_samples=300,noise=0.3, random_state=6) #random_state : seed du générateur 

maxdepth=1
M=20
alphas, models = Adaboost(Y,c,maxdepth,M)
accuracy=Adaboost_accuracy(alphas, models, Y, c)


fig=plt.figure(figsize=(17,10))
fig.suptitle('Arbres successivement apris', size=20)
for i in range(M):
    plt.subplot(5,4,i+1)
    plt.title('arbre %i' %(i+1))
    frontieres(models[i], Y)
    
fig=plt.figure(figsize=(17,10))
fig.suptitle("Décisions successives d'Adaboost :", size=20)
for i in range(M):
    plt.subplot(5,4,i+1)
    plt.title('itération %i' %(i+1))
    frontieres_Adaboost(alphas[:i+1], models[:i+1], Y)
```

**On observe bien les frontières de décision associées à chaque arbre et associées à chaque étape de l'algorithme Adaboost.**

**On observe plusieurs occurences de certaines de ces frontières au fur et à mesures des itérations de l'algorithme.**

## Exercice 2 : Stacked NBCs

*Dans cet exercie, chaque classifieur de base verra des données appartenant à une région spécifique de l’input
space. Les classifieurs de base n’auront donc pas accès à un dataset i.i.d. Le classifieur de second niveau aura
lui accès à 20% de chaque dataset local. L’union de ces fractions de datasets constituera un échantillon i.i.d.
permettant au comité d’obtenir de bonnes performances globales.*

*On considérera qu’un classifieur choisit une unique classe : fm (x) ∈ C où C est l’ensemble des classes. On fixe
la taille du comité à M = 4. Le classifieur de second niveau sera aussi un Naive Bayes Classifier (NBC), mais
ce dernier s’appuiera sur une distribution multinomiale. Pour les classifieur de premier niveau, la distribution
sous-jacente sera Gaussienne car les inputs x sont des vecteurs de $R^2$*.

*Nous utiliserons les implémentations du NBC fournies par le module `sklearn`.*


### Question 1 : 
*Mettez en préambule de votre code les instructions suivantes :*

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.naive_bayes import GaussianNB, MultinomialNB

from sklearn.model_selection import train_test_split
import seaborn as sns
```

### Question 2 :
*Nous allons commencer par générer le dataset synthétique qui nous servira de terrain de jeu pour cet
exercice. Pour ce faire, vous appellerez la fonction `gen_blobs` fourni par l’enseignant. Vous spécifierez la
taille souhaitée pour le dataset en argument de cette fonction, à savoir n = 300.*

**Fonction `gen_blobs` :**

```python
def gen_blobs(n, beta=0.5, random_state=None):
    """
    This function generates a dataset from the following generating process: 
    4 gaussian 2D distributions centered on each corner of a centered square
    whose side length is 4. Each gaussian has unit variance
    The diagonal blobs generates example of class n° 0. The anti-diagonal blobs 
    generates examples of class n°1 and n°2 respectively.
    """
    X1, y1 = make_blobs(shuffle=False,n_samples=int(n*beta),n_features=2, centers=np.asarray([[-1.5,-1.5],[-1.5,1.5] ]), random_state=random_state)
    X2, y2 = make_blobs(shuffle=False,n_samples=int(n*(1-beta)),n_features=2, centers=np.asarray([[-1.5,-1.5],[-1.5,1.5] ]), random_state=random_state)
    X3, y3 = make_blobs(shuffle=False,n_samples=int(n*beta),n_features=2, centers=np.asarray([[1.5,1.5],[1.5,-1.5] ]), random_state=random_state)
    X4, y4 = make_blobs(shuffle=False,n_samples=int(n*(1-beta)),n_features=2, centers=np.asarray([[1.5,1.5],[1.5,-1.5] ]), random_state=random_state)
    X = np.vstack((X1[np.where(y1==0)],X2[np.where(y2==1)],X3[np.where(y3==0)],X4[np.where(y4==1)]))
    y = np.hstack((y1[np.where(y1==0)],y2[np.where(y2==1)],y3[np.where(y3==0)],y4[np.where(y4==1)]))
    perm = np.random.permutation(len(y))
    X = X[perm]
    y = y[perm]
    return (X,y)
```

**Génération du dataset :**

```python
(Y,c) = gen_blobs(300, random_state=None) 
```

**Affichage du dataset :**

```python
plt.scatter(Y[c==0][:,0], Y[c==0][:,1])
plt.scatter(Y[c==1][:,0], Y[c==1][:,1],color='r')
plt.title('dataset')
plt.show()
```

### Question 6: 

*Construisez des vecteurs c de taille 4, tels que cm = fm (x), où l’input x est un vecteur des 20% de
données mises de côté. Instanciez un NBC en appelant le constructeur de la classe MultinomialNB() que
vous entraînerez grâce aux vecteurs c et leurs étiquettes associées y.*

```python
def vector_C(models=models, Y2=Y2, c2=c2):
    #chaque ligne de C correspond un vecteur c tel que définit plus haut
    C=np.array([model.predict(Y2) for model in models]).T
    model_MNB=MultinomialNB()
    model_MNB.fit(C, c2)
    return C, model_MNB
```

### Question 7 :

*Générez un ensemble de test de même taille que celui initial (n = 300) et calculez les scores de classifications obtenus par les quatre classifieurs ainsi que par le comité.*

```python
(Y,c) = gen_blobs(300, random_state=5) 

def Scores_clf_comite(models,Y, c):
    Scores=[]
    for i in range(4):
        model=models[i]
        Scores.append(model.score(Y,c))
        
    C, model_MNB=vector_C(models, Y,c)
    Scores.append(model_MNB.score(C,c))    
    return Scores
```

**On affiche les scores des 4 classifieurs et du comité :**

```python
Scores=Scores_clf_comite(models, Y, c)
for i in range(4):
    print('classifieur {i} : {result}'.format(i=i+1,result=Scores[i]))
print('Score du comité : {result:.2f}'.format(result=Scores[4]))
```

**On observe que le score du comité est bien supérieur à chacun des scores des quatre classifieurs, ce qui est satisfaisant et montre que cette méthode fonctionne.**
