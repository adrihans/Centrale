## Calcul de la valeur de la densité de probabilité d'un mélange de gaussienne en un point
### Les paramétres de la fonction gm_pdf
<ul>
<li>x : le point où l'on calcule la valeur de la densitè</li>
<li>mu : le vecteur des moyennes des gaussiennes</li>
<li>sigma : le vecteur des écart-types des gaussiennes</li>
<li>p : le vecteur des probabilitès de la loi multinomiale associèe</li>
</ul>

```python
import numpy as np
from scipy.stats import norm
from scipy.stats import uniform
import matplotlib.pyplot as plt
```

```python
def gm_pdf(x, mu, sigma, p):
    #Initialisation de la variable de sortie
    resultat = 0.0
    #Contrôle de la cohérence des paramètres d'entrée
    #Le vecteur de moyenne doit avoir la même longueur que le vecteur p
    if len(mu) != len(p):
        print('Erreur de dimension sur la moyenne')
    # Le vecteur des écart-types doit avoir la même longueur que le vecteur p
    elif len(sigma) != len(p):
            print('Erreur de dimension sur les écarts-types')
    else:
    # Calcul de la valeur de la densité
        for i in range(0, len(p)):
            resultat = resultat + p[i] * norm.pdf(x, mu[i], sigma[i])
    return resultat
```

## Génération de nombre aléatoire suivant un mélange de gaussienne
Dans un mélange de gaussienne chaque densité de probabilité correspondant à une loi normale est pondérée par un coefficient plus petit que un.
La somme de tous ces coefficients est égale à un. Une variable aléatoire qui suit une loi normale suit donc une des lois normales du mélange avec une probabilité égale au coefficient de pondération de cette même loi normale. Pour générer un échantillon suivant une loi normale, il faut donc procéder en deux étapes :
<ol>
<li>Tirer aléatoirement un nombre entre 1 et N (le nombre de gaussiennes du mélange) suivant une loi mumltinomiale définie par les coefficients du mélange.</li>
<li>Une fois ce nombre obtenu, on génère le nombre suivant la loi normale associée.</li>
</ol>
### Les paramètres de la fonction gm_rnd
<ul>
<li>mu : le vecteur des moyennes des gaussiennes</li>
<li>sigma : le vecteur des écart-types des gaussiennes</li>
<li>p : le vecteur des probabilités de la loi multinomiale associée</li>
</ul>

```python
def gm_rnd(mu, sigma, p):
    # Initialisation de la variable de sortie
    resultat = 0.0
    #Contrôle de la cohérence des paramètres d'entrée
    #Le vecteur de moyenne doit avoir la même longueur que le vecteur p
    if len(mu) != len(p):
        print('Erreur de dimension sur la moyenne')
    # Le vecteur des écart-types doit avoir la même longueur que le vecteur p
    elif len(sigma) != len(p):
            print('Erreur de dimension sur sur les écarts-types')
    else:
    #Génération de l'échantillon
    # On échantillonne suivant une loi uniforme sur [0,1]
        u = uniform.rvs(loc = 0.0, scale = 1.0, size = 1)
    # % Chaque test suivant permet de définir un intervalle sur lequel la
    # probabilité d'appartenance de la variable uniforme est égale à l'une des
    # probabilités définie dans le vecteur p. Lorsque u appartient à l'un de
    # ces intervalles, c'est équivalent à avoir générer une variable aléatoire
    # suivant l'un des éléments de p. Par exemple, pour le premier test
    # ci-dessous, la probabilité que u appartienne à l'intervalle [0,p[0][ est
    # égale à p[0] puisque u suit une loi uniforme. Donc si u appartient à
    # [0,p[0][ cela est équivalent à avoir tirer suivant l'événement de probabilité p[0].
        if u < p[0]: # On test si on a généré un événement de probabilité p[0]
            resultat = sigma[0] * norm.rvs(loc = 0, scale = 1, size = 1) + mu[0]
            # Pour générer suivant une loi normale quelconque, il suffit de multiplier
            # une variable normale centrée réduite (moyenne nulle et écart-type égal à 1)
            # par l'écart-type désité et d'additionner la moyenne désirée au produit précédent.
        for i in range(1, len(p)):
            if (u > np.sum(p[0:i])) and (u <= np.sum(p[0:i+1])): # On test si on a généré
                # un événement de probabilité p[i]
                resultat = sigma[i] * norm.rvs(loc = 0.0, scale = 1.0, size = 1) + mu[i]
                # Pour générer suivant une loi normale quelconque, il suffit de multiplier
                # une variable normale centrée réduite (moyenne nulle et écart-type égal à 1)
                # par l'écart-type désité et d'additionner la moyenne désirée au produit précédent.
    return resultat
```

### L'algorithme EM

Pour pouvoir utiliser les donnees galaxy, on execute cette ligne de code :

```python
donnees=donnees_galaxy
```

Pour avoir de bonnes valeurs de départ, on prend pour les valeurs $\mu$_em,$\sigma$_em et $\alpha$_em les valeurs que l'on a trouvé plus haut via les visualisations du mélange de gaussienne.

```python
#Nombre maximum d'itérations 
nbMaxIterations = 70
#Valeurs initiales
mu_em = np.array([10000, 20000, 22000, 24000, 33000])
sigma_em = np.array([300, 500, 500, 500, 300])
alpha_em = np.array([0.050, 0.40, 0.300,0.2, 0.05])

nbIteration = 1 #Initialisation de la variable d'arrêt
nbComposante = len(alpha_em) #Nombre de composantes du mélange
nbDonnees = len(donnees)  #Nombre de données
p = np.zeros(shape=(nbComposante, nbDonnees))
#Déclaration et initialisation de la matrice qui va contenir les probabilités
#p(k|x,theta_courant)

#On ajoute la ligne suivante pour initialiser le nombre d'échantillons :  
nbEchantillon = 1000
alpha_em_new = alpha_em
sigma_em_carre_new = sigma_em
mu_em_new = mu_em
donneesP = np.zeros(shape=(nbEchantillon))

while nbIteration < nbMaxIterations:
    for n in range(0, nbDonnees, 1):
        for k in range(0, nbComposante, 1):
            p[k, n] = alpha_em[k] * norm.pdf(x = donnees[n], loc = mu_em[k], scale = sigma_em[k])
        p[:, n] = p[:, n] / np.sum(p[:, n])
    for k in range(0, nbComposante, 1):
        alpha_em_new[k] = np.sum(p[k,:]) / nbDonnees
        for n in range(0, nbDonnees, 1):
            donneesP[n] = donnees[n] * p[k, n]
        mu_em_new[k]  = np.sum(donneesP) / np.sum(p[k, :])
        for n in range(nbDonnees):
            donneesP[n] = ((donnees[n] - mu_em_new[k]) ** 2) * p[k, n]
        sigma_em_carre_new[k] = np.sum(donneesP) / np.sum(p[k, :])
    mu_em = mu_em_new
    sigma_em = np.sqrt(sigma_em_carre_new)
    alpha_em = alpha_em_new
    nbIteration = nbIteration + 1
```

```python
print('Les paramètres estimés sont : ')
print('Moyennes des composantes du mélange', mu_em)
print('Ecrat type des composantes du mélange', sigma_em)
print('Probabilités des composantes du mélange', alpha_em)
```

```python
print('La somme des probabilités des composantes du mélange vaut : ', np.sum(alpha_em))
```

La somme est 1 donc il n'y a pas de problème.

Même si ces résultats se rapprochent assez bien de la distributions des données magré seulement 5 gaussiennes, nous allons maintenant essayé d'avoir des résultats plus justes en testant sur plusieurs nombres de gaussiennes considérées : 

## Essai avec plusieurs nombres de gaussiennes : 

On défini ci-dessous une fonction algorithme_EM pour permettre de tester plusieurs paramètres sans avoir à réécrire le même code à chaque fois : 

```python
def algorithme_EM(mu_em, sigma_em, alpha_em, nbMaxIterations,donnees):
    
    nbIteration = 1 #Initialisation de la variable d'arrêt
    nbComposante = len(alpha_em) #Nombre de composantes du mélange
    nbDonnees = len(donnees)  #Nombre de données
    p = np.zeros(shape=(nbComposante, nbDonnees))
    #Déclaration et initialisation de la matrice qui va contenir les probabilités
    #p(k|x,theta_courant)

    #On ajoute la ligne suivante pour initialiser le nombre d'échantillons :  
    nbEchantillon = 100
    alpha_em_new = alpha_em
    sigma_em_carre_new = sigma_em
    mu_em_new = mu_em
    donneesP = np.zeros(shape=(nbEchantillon))

    while nbIteration < nbMaxIterations:
        for n in range(0, nbDonnees, 1):
            for k in range(0, nbComposante, 1):
                p[k, n] = alpha_em[k] * norm.pdf(x = donnees[n], loc = mu_em[k], scale = sigma_em[k])
            p[:, n] = p[:, n] / np.sum(p[:, n])
        for k in range(0, nbComposante, 1):
            alpha_em_new[k] = np.sum(p[k,:]) / nbDonnees
            
            for n in range(0, nbDonnees, 1):
                donneesP[n] = donnees[n] * p[k, n]
            mu_em_new[k]  = np.sum(donneesP) / np.sum(p[k, :])
            
            for n in range(nbDonnees):
                donneesP[n] = ((donnees[n] - mu_em_new[k]) ** 2) * p[k, n]
            sigma_em_carre_new[k] = np.nansum(donneesP) / (np.nansum(p[k, :])+1e-15)

        mu_em = mu_em_new
        sigma_em = np.sqrt(sigma_em_carre_new)
        alpha_em = alpha_em_new
        nbIteration = nbIteration + 1
        
    return mu_em, sigma_em, alpha_em
```

On va maintenant faire une boucle for sur un nombre de gaussiennes et afficher le résultat pour chaque itération : 

```python
nbGaussiennesMax=15
nbMaxIterations=80
```

```python
for i in range(6,nbGaussiennesMax):
    
    mu_em = np.array([10000, 17000,27000,33000] + [18000 + 11000*((k)/i) for k in range(0,i-4)])
    sigma_em = np.array([500]*i)
    alpha_em = np.array([1/i]*i)
    
    pointPdf = gm_pdf(x, mu_em, sigma_em, alpha_em)
    mu_em, sigma_em, alpha_em = algorithme_EM(mu_em,sigma_em,alpha_em,nbMaxIterations,donnees_galaxy)
    pointPdfEstime = gm_pdf(x, mu_em, sigma_em, alpha_em)
    
    print('pour %i gaussiennes ' %i)
    print('Les paramètres estimés sont : ')
    print('Moyennes des composantes du mélange', mu_em)
    print('Ecrat type des composantes du mélange', sigma_em)
    print('Probabilités des composantes du mélange', alpha_em)
    
    plt.figure(figsize=(15,5))
    
    plt.subplot(1,2,1)
    plt.plot(x, pointPdf, 'r-', label = 'Originale')
    plt.title('Densite de melange')
    plt.xlabel('x')
    plt.ylabel('pdf')
    plt.plot(x, pointPdfEstime, 'b-', label = 'Estimée')
    plt.legend(loc='upper left', shadow=True, fontsize='x-large')

    plt.subplot(1,2,2)
    plt.title('Densite de melange et histogramme des données galaxy')
    plt.hist(donnees_galaxy, bins=30,density=True,color='green',label='histogramme des donnees galaxy')
    plt.xlabel('x')
    plt.ylabel('pdf')
    pointPdfEstime = gm_pdf(x, mu_em, sigma_em, alpha_em)
    plt.plot(x, pointPdfEstime, 'b-', label = 'Estimée')
    plt.legend(loc='upper left', shadow=True, fontsize='x-large')


    plt.show()
```

# Conclusion : 

Nous avions une erreur sur `np.sum(donneesP)` par exemple, car il y avait des nans, donc nous avons utilisé np.nansum au lieu de np.sum et nous avons empécher la possibilité de division par 0.
 
**Nous observons via ces représentations graphiques que :**

- De manière générale, et comme attendu, plus le nombre de gaussiennes est important, plus la courbe estimée se rapproche de la distribution des données galaxy.

- Cependant, on observe tout de même une très grande sensibilité à peu d'évènements. (pic vers 27000 apparaissant entre 6 et 7 gaussiennes par exemple).
Ceci est sûrement dû au fait que l'on attribue la même valeur de probabilité pour toutes les gaussiennes peu importe leurs positions initialement, et que le centre (la moyenne) d'une gaussienne se trouve à l'endroit de ce groupe de données.
Nous observons d'ailleurs que sur le nombre de gaussiennes considéré, les meilleurs résultats sont finalement ceux obtenus pour 13 gaussiennes-en tout cas visuellement- à cause de l'apparition de deux pics juste avant 25000 qui ont une 'intensité' bien plus importante que la distribution des données à cet endroit.
Par contre, cette même constatation est sûrement dû aussi au peu de gaussiennes considérées. On peut en effet s'attendre à avoir des résultats très très proche étant donné le comportement de l'algorithme, pour un très grand nombre de gaussiennes (ces mêmes pics 'étonnants' ne seraient alors pas d'une intensité aussi importante) 
- On a vu aussi que l'algorithme peut être instable en fonction du nombre de gaussienne considéré  et des conditions initiales(il ne retourne que des nans dans certains cas) et qu'il dépend très fortement de ces conditions initiales. 


Il serait alors peut-être judicieux de tester plusieurs formes de conditions initiales (ou d'avoir un algorithme permettant de déterminer les meilleures en en testant plusieurs directement) afin d'avoir des résultats plus proches des données considérées.
