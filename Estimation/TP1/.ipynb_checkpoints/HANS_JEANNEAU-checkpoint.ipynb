## I : A partir de l’inverse généralisée de la fonction de répartition

Ici, la fonction de répartition de X est égale à : 

\begin{align*}
F(x) = 1 - exp(-\lambda x)
\end{align*}

1) Après calcul, l'inverse généralisée de F est égale à  : 
\begin{align*}
F^{-1}(u) = - \frac{ln(1-u)}{\lambda}
\end{align*}

2) On utilise la méthode de l'inverse généralisée pour simuler 10 000 échantillons issus d'une loi exponentielle avec $\lambda = 2$. 

```python
Lambda=2
nbEchantillon = 10000
U=uniform.rvs(loc=0.0,scale=1.0,size=nbEchantillon)
X=(-1/Lambda)*np.log(1-U)
```

```python
#On affiche le résultat: 
fig=plt.figure(figsize=(5,5))
plt.scatter(np.arange(0,10000),X,color='r')
plt.show()
```

3) On trace sur une autre figure la courbe de densité de probabilité normalisée de X :

```python
x_dens=np.arange(0,5,0.01)
densite_X=Lambda*np.exp(-Lambda*x_dens)
```

```python
fig=plt.figure()
plt.plot(x_dens,densite_X)
plt.show()
```

4) On trace l'histogramme normalisé des échantillons de la première question.

```python
plt.hist(X,bins=50,normed=True)
plt.plot(x_dens,densite_X)
plt.show()
```

**On peut calculer certaines propriétés des 10 000 échantillons précédemment simulés : **

```python
print('Moyenne Empirique :' , np.mean(X))
print('Variance Empirique :' , np.var(X))
```

## II : Méthode Accept-Reject

```python
# Définition des fonctions et constantes utiles
X = np.linspace(-3, 3, 100)
f = norm.pdf
f_dist = norm.rvs
g = cauchy.pdf
g_dist = cauchy.rvs
M = np.sqrt(2*np.pi/np.e)
```

Voici l'allure des deux fonctions f et g avec lesquelles nous allons travailler.

```python
plt.plot(X, f(X), label='fonction $f$')
plt.plot(X, M*g(X), label='fonction $M*g$')
plt.legend()
_ = plt.xlabel('$x$')
_ = plt.ylabel('$y$')
_ = plt.title('Fonctions $f$ et $M*g$')
```

Nous implémentons maintenant la méthode Accepte-rejecte comme demandé dans l'énoncé.

```python
# Accept-reject Method
def Accepte_rejecte(f, g, g_dist, M, nb_it=1000):
    res = []
    for i in range(nb_it):
        u = uniform.rvs(loc=0, scale=1)
        x = g_dist()
        while (f(x) / (M * g(x)) < u):
            u = uniform.rvs(loc=0, scale=1)
            x = g_dist()
        res.append(x)
    return res
```

Nous appliquons maintenant cette méthode à la loi centrée réduite avec $M = 5$ et 10 000 échantillons

```python
res = Accepte_rejecte(f, g, g_dist, 5, nb_it=10000)
```

```python
plt.plot(X, f(X), label='Loi normale')
_ = plt.hist(res, density=True, bins=X, label='Echantillon')
_ = plt.xlabel('$x$')
_ = plt.ylabel('$y$')
_ = plt.title('echantillonnage de la loi normale selon la méthode Accepte-rejecte')
plt.legend()
```

Le résultat est satisfaisant, l'échantillon suit bien la loi de probabilité donnée en ayant un bruit assez faible

```python
res2 = Accepte_rejecte(norm(5, 1).pdf, g, g_dist, 5, nb_it=10000)
```

```python
plt.plot(X+5, norm(5, 1).pdf(X+5), label='Loi normale de moyenne 5')
plt.plot(X+5, g(X+5), label='g')
_ = plt.hist(res2, density=True, bins=X+5, label='Echantillon', fc=(1, 0, 0, 0.5))

ttt = norm.rvs(loc=5, scale=1, size=10000)
_ = plt.hist(ttt, density=True, bins=X+5, fc=(0, 1, 0, 0.5), label='ground truth')

_ = plt.xlabel('$x$')
_ = plt.ylabel('$y$')
_ = plt.title('echantillonnage $\mathcal{N}(5, 1)$ selon la méthode Accepte-rejecte')
plt.legend()
```

En reproduisant la méthode précédente pour echantillonner $\mathcal{N}(5, 1)$, nous n'obtenons pas le résultat attendu. En effet, nous n'avons pas modifié la fonction g, nous n'avons plus $f(x) < M*g(x)$ sur la plage d'échantillonnage. Il faudrait translater la fonction g de 5 pour retrouver un résultat convenable.

## V : Echantilloner suivant une loi de Bernoulli : 
1) On génère 1000 échantillons suivant une loi de Bernoulli de paramètre p=0.7 puis on trace le résultat : 

```python
p=0.7
nbEchantillon = 1000
#On génère 1000 nombres pseudo-aléatoires uniforme entre 0 et 1 : 
U=uniform.rvs(loc=0.0,scale=1.0,size=nbEchantillon)
X=[]
for u in U:
    if u<p:
        X.append(0)
    else:
        X.append(1)
```

```python
fig=plt.figure()
plt.hist(X)
plt.title('histogramme, suivant loi de Bernoulli avec p=0.7')
plt.show()
```

2) On calcule la fréquence du nombre de zéros générés : 

```python
print('fréquence du nombre de zéros générés : ' , X.count(0)/nbEchantillon)
```

```python
print('Moyenne empirique : ', np.mean(X))
print('Variance empirique : ', np.var(X))
```

On retrouve le paramètre de la loi de Bernoulli à $10^{-3}$ près. 

##  VI : Les méthodes Monte Carlo Markov Chain : MCMC

On implémente ici l'algorithme de Metropolis-Hastings tel que décrit dans l'énoncé (Algorithme 8)

L'algorithme prend en entrée la fonction a échantilloner dont on connait la fonction de distribution p, l'element initial x0, la fonction de proposition prop, ainsi que le nombre de sample et le burn-in

Afin de prendre directement en entrée la fonction de proposition, nous avons écrit `y = x + prop()`. Cela fonctione car dans nos deux cas,
$f \sim \mathcal{U}(x+a, x+b) \Rightarrow f \sim x + \mathcal{U}(a, b)$ 
and  
$f \sim \mathcal{N}(x, \sigma) \Rightarrow f \sim x + \mathcal{N}(0, \sigma)$

```python
def MCMC(p, x0, prop, n_samples=100, burn_in=0):
    """
    Algorithme de Metropolis-Hastings utlisant une loi de proposition uniforme.
    p: fonction a echantilloner
    x0: element initial
    n_samples: Nombre final d'échantillons
    burn_in: burn-in
    """
    # Initialisation
    sample_list = [x0]
    x = x0
    # compteur d'échantillons pour le burn-in
    n_sampled = 0

    while len(sample_list) < n_samples:
        # choisi un candidat dans U(x-uni, x+uni) pour la première partie,
        # ou N(0, s) pour la seconde
        y = x + prop()
        rnd = np.random.rand()
        try:
            rho = min(1, p(y) / p(x))
        except (OverflowError, ValueError) as error:
            continue
        n_sampled += 1

        # Accepte le nouveau candidat
        if rho > rnd:
            x = y  

        # Ne pas ajouter les echantillons du burn-in
        if n_sampled > burn_in:
            sample_list.append(x)

    return np.array(sample_list)
```

Nous definissons ensuite une fonction affichant la l'echantillonnage effectué ainsi que la distribution cible :

```python
def printsamples(samples, pdf_f, title="Sampling normal distribution (MCMC)", bins=15, labels=True):
    labelhist = 'histogram of samples' if labels else ''
    labelplot = 'function distribution' if labels else ''
    plt.hist(samples, density=True, bins=bins, label=labelhist)
    X = np.linspace(min(samples), max(samples), 100)
    Y = [pdf_f(x) for x in X]
    plt.plot(X, Y, label=labelplot)
    plt.xlim(min(samples), max(samples))
    plt.title(title)
    plt.xlabel(r'$x$')
    plt.ylabel(r'$y$')
    plt.legend()
```

Par exemple, voici l'echantillonnage de la loi uniforme $\mathcal{N}(0, 1)$ en utilisant comme fonction de proposition $\mathcal{U}(-1, 1)$, avec 20 000 echantillons et un burn-in de 500

```python
samples1 = MCMC(norm(0, 1).pdf, np.random.randn(), uniform(-1, 2).rvs, 20000, 500)
```

```python
printsamples(samples1, norm(0, 1).pdf, title="Sampling normal distribution (MCMC) with $U(-1, 1)$", bins=40)
```

En comparaison, voici l'echantillonnage de la loi uniforme $\mathcal{N}(0, 1)$ en utilisant comme fonction de proposition $\mathcal{U}(-20, 20)$, avec 20 000 echantillons et un burn-in de 1000

```python
samples2 = MCMC(norm(0, 1).pdf, np.random.randn(), uniform(-20, 40).rvs, 20000, 1000)
```

```python
printsamples(samples2, norm(0, 1).pdf, title="Sampling normal distribution (MCMC) with $U(-20, 20)$", bins=40)
```

Nous observons qu'en augmentant significativement le support de la fonction de proposition, le résultat obtenu est de moins bonne qualité, car plus bruité: l'échantillonnage est moins fidèle à la fonction de distribution. En effet, un plus grand support de la fonction implique une plus grand volatilité dans les valeurs que peuvent prendre les candidats

## 6.2

Enfin, interessons-nous à une autre fonction, plus complexe car comprenant deux modes (deux "pics").

```python
def f(x, a1=1, a2=2, mu1=10, mu2=-5, p=0.3):
    exp1 = p/(2*a1) * exp(-abs((x-mu1)/a1))
    exp2 = (1-p)/(2*a2) * exp(-abs((x-mu2)/a2))
    return exp1 + exp2
```

```python
X = np.linspace(-20, 20, 100)
Y = [f(x) for x in X]
plt.plot(X, Y)
```

Nous voulons échantillonner cette fonction à l'aide d'une fonction de proposition gaussienne. 

Essayons différentes valeurs de $\sigma$: 

```python
plt.subplot(221)
testsamples = MCMC(f, np.random.randn(), norm(0, 1).rvs, 2000, 100)
printsamples(testsamples, f, title="Sampling with $\mathcal{N}(0, 1)$", bins=40, labels=False)
plt.subplot(222)
testsamples = MCMC(f, np.random.randn(), norm(0, 2).rvs, 2000, 100)
printsamples(testsamples, f, title="Sampling with $\mathcal{N}(0, 2)$", bins=40, labels=False)
plt.subplot(223)
testsamples = MCMC(f, np.random.randn(), norm(0, 4).rvs, 2000, 100)
printsamples(testsamples, f, title="Sampling with $\mathcal{N}(0, 4)$", bins=40, labels=False)
plt.subplot(224)
testsamples = MCMC(f, np.random.randn(), norm(0, 8).rvs, 2000, 100)
printsamples(testsamples, f, title="Sampling with $\mathcal{N}(0, 8)$", bins=40, labels=False)
plt.tight_layout()
```

Nous remarquons ici que pour $\sigma = 1$ et $\sigma = 2$, les echantillons tendent à rester bloqués dans un mode et ne suivent pas fidèlement la distribution voulue. Les échantillons ne peuvent pas "sauter" d'un mode à l'autre car le gap est trop grand devant l'écart type de la fonction de proposition, il est donc peu probable qu'un nouvel échantillon change de mode.

A l'inverse, pour $\sigma = 8$, les deux modes sont représentés equitablement et il n'y a pas de problème de pasage d'un mode à l'autre. Cependant, un trop grand écart-type dans la fonction de proposition rend les échantillons trop volatiles, tout comme un trop grand support de cette même fonction comme vu précédemment. L'echantillonnage est trop bruité et peu fidèle à la fonciton de distribution $f$, $\mathcal{N}(0, 8)$ n'est donc pas un bon choix de fonction de proposition.

Il semble donc raisonnable de choisir $\sigma = 4$. Une fois ce choix établi sur des échantillons de taille moyenne, pour des raisons de tempsd de calcul, reproduisons l'experience avec 20 000 échantillons et un burn-in de 1 000.

```python
samples3 = MCMC(f, np.random.randn(), norm(0, 4).rvs, 20000, 500)
```

```python
printsamples(samples3, f, title="Sampling function f (MCMC) with $N(0, 1)$", bins=40)
```

Le résultat obtenu est satisfaisant. L'echantillonnage suis fidèlement la fonction de distribution donnée, et les deux modes sont bien représentés. Cela valide notre choix de $\sigma$
