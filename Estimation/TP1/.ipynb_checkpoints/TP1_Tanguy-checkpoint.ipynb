## II. MÃ©thode Accept-Reject

```python
import numpy as np
from scipy.stats import uniform, norm, cauchy
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-3, 3, 100)
f = norm.pdf
f_dist = norm.rvs
g = cauchy.pdf
g_dist = cauchy.rvs
M = np.sqrt(2*np.pi/np.e)
```

```python
plt.plot(X, f(X))
plt.plot(X, M*g(X))
```

1.

```python
# Accept-reject Method
def Accepte_rejecte(f, g, g_dist, M, nb_it=1000):
    res = []
    for i in range(nb_it):
        u = uniform.rvs(loc=0, scale=1)
        x = g_dist()
        while (f(x) / (M * g(x)) < u):
            u = uniform.rvs(loc=0, scale=1)
            x = g_dist()
        res.append(x)
    return res
```

```python
res = Accepte_rejecte(f, g, g_dist, 5, nb_it=10000)
```

```python
plt.plot(X, f(X))
_ = plt.hist(res, density=True, bins=X)
```
