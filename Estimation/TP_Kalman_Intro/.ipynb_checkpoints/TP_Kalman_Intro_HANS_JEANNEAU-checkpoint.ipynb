# Kalaman filter :

## Estimating a constant :

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Kalman(N, R=1, Q=1e-5, C=[1, 5, -5]):
    #Initialization :
    Lz=[]
    Lx_hat=[]
    Lt=[]
    x_hat=-2
    P=1
    x=np.random.normal(x_hat,P)
    t=0
    while t<N:
        val = C[int(len(C)*t/N)]
        
        P=P+Q
        t+=1
        x=val+np.random.normal(0,Q)
        z=np.random.normal(x,R)
        K=P*(1/(R+P))
        x_hat=x_hat+K*(z-x_hat)        
        P=P-K*P
        
        Lx_hat.append(x_hat)
        Lz.append(z)
        Lt.append(t)
    return Lt, Lx_hat, Lz
```

**$$Q=10^{-5} :$$**

```python
def plot_Kalman(N, R=[1], Q=[1e-5], C=[1, 5, -5], legend=True, figsize=(10, 5), ylim=None):
    K = len(R)
    M = len(Q)
    plt.figure(figsize=figsize)

    for i in range(K):
        for j in range(M):
            plt.subplot(K, M, i*M+j+1)
            Lt, Lx_hat, Lz = Kalman(N, R=R[i], Q=Q[j], C=C)
            vals = [C[int(len(C)*t/len(Lt))] for t in range(len(Lt))]
            plt.plot(Lt, vals, color='r', label="constante")
            plt.plot(Lt, Lx_hat, color='g', label="estimation")
            plt.scatter(Lt, Lz, color='b', alpha=0.1,label= "mesures")
            plt.title('$R={1:.0e}{0}, Q={2:.0e}{0}$'.format('}', R[i], Q[j]).replace('e', '^{'))
            if legend:
                plt.legend()
            if ylim is not None:
                plt.ylim(ylim)
    plt.tight_layout()
    plt.show()
```

```python
plot_Kalman(1000, C=[1])
```

The Kalman filter is correctly estimating the constant, even if the constant is significantly greater than the initial value as below:

```python
plot_Kalman(1000, R=[1], Q=[1e-5], C=[100])
```

The value of R controls the noise in the measurement of the data, and the value of Q controls the noise in the data itself. This can be observed when trying different values of R or Q.

```python
plot_Kalman(1000, R=np.logspace(-0.5, 0.5, 3), Q=np.logspace(-5, -2, 4), C=[1], legend=False, figsize=(15, 15), ylim=(-2, 5))
```

We observe that given a value of R, the greater the Q, the greater the predicted signal oscillations will be around the constant value to estimate.

Similarly, given a value of Q, the greater the R, the greater the predicted signal oscillations will be around the constant value to estimate.

When estimating a constant, it appears that a smaller value of Q and R are desirable, since it limitates the variability around the value to estimate. However, we will apply the same reasoning to the estimation of a square signal and observe the advantages of a higher Q.

## Tracking a square signal :

```python
plot_Kalman(1000, C=[1, 5, -5])
```

With the same parameters as in the first plot, we notice that the Kalman Filter cannot estimate accurately a square signal. More precisely, there is a great transition time that prevent a quick change in the estimated value.

```python
plot_Kalman(1000, R=[1, 0.1], C=[1, 5, -5])
```

A diminution of R by one order of magnitude has a visible yet not sufficient effect on the response time of the filter.

```python
plot_Kalman(1000, Q=[1e-4, 1e-5], C=[1, 5, -5])
```

Similarly, a decreasing Q leads to an improvement in the response time. One order of magnitude is however barely sufficient.

```python
plot_Kalman(1000, Q=[1e-1, 1e-5], C=[1, 5, -5], ylim=(-6, 6))
```

Testing two extreme values of Q highlights the trade-off between rapidity of the response time and accuracy / variability of the stationary state estimated value.

```python
plot_Kalman(2000, Q=[1e-3, 1e-4, 1e-5], C=1+1/2*(-1)**np.array(range(10)))
```

We can also try a different square signal. Here, we notice that there is no "sweet spot" in the accuracy-rapidity trade-off that allow an accurate and fast estimation of the square signal.
